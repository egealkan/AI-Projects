# Conclusion and Comparison Between Models for Stock Forecasting

In this analysis, we compared the performance of three different models—**LSTM (Long Short-Term Memory)**, **PyCaret**, and **ARIMA (Auto-Regressive Integrated Moving Average)**—to forecast stock prices for **Microsoft (MSFT)** and **Apple (AAPL)**. Below are the key insights from our analysis:

### 1. **LSTM (Long Short-Term Memory)**
   - **Performance**: LSTM demonstrated strong performance across both AAPL and MSFT datasets.
     - For **AAPL**, the model achieved a high **R-squared** of **0.9956**, indicating that it was able to explain a large proportion of variance in the stock prices. 
     - For **MSFT**, the **R-squared** value of **0.9947** was also quite impressive, suggesting that LSTM captures the time-series patterns well for MSFT stock as well.
   - **Error Metrics**: The **MSE** values of **0.3415** (AAPL) and **1.7123** (MSFT) show that LSTM was able to minimize error and provide reasonable predictions.

### 2. **PyCaret**
   - **Performance**: PyCaret delivered the best overall performance, with an almost perfect fit for both stocks.
     - The **R-squared** values were **0.99997** for **AAPL** and **0.9999** for **MSFT**, demonstrating exceptional model accuracy.
   - **Error Metrics**: The **MSE** for AAPL was **0.0044** and for MSFT was **0.0393**, showing very low error rates. PyCaret was especially effective in minimizing errors while predicting stock prices for both companies.

### 3. **ARIMA (Auto-Regressive Integrated Moving Average)**
   - **Performance**: ARIMA showed the weakest results, particularly for **AAPL**, where the **R-squared** was negative (**-0.0091**), indicating a poor fit.
     - For **MSFT**, ARIMA performed slightly better, with an **R-squared** of **0.9947**, indicating a moderate fit to the data.
   - **Error Metrics**: The **MSE** value for AAPL was **6.0534**, which was quite large compared to the other models. For MSFT, **MSE** was **1.7123**, which still indicates significant error.

### Comparison Summary:
- **Best Performance**: **PyCaret** outperformed both **LSTM** and **ARIMA**, with the highest **R-squared** values and lowest error metrics, making it the most effective model for stock forecasting.
- **LSTM**: While not as good as PyCaret in terms of fit, LSTM provided solid predictions for both stocks with reasonable **R-squared** and **MSE** values.
- **ARIMA**: ARIMA was the least effective model, with poor performance for **AAPL** and only marginal improvement for **MSFT**.

### Final Thoughts:
- For stock price forecasting, **PyCaret** should be the preferred model given its near-perfect fit and minimal error.
- **LSTM** is a reliable second choice, particularly if interpretability and time-series capturing are important.
- **ARIMA** does not appear suitable for predicting stock prices in this instance, as it struggled to model the data effectively.

The insights gathered through this analysis will help guide the selection of the most appropriate model for future stock price predictions for both **Microsoft (MSFT)** and **Apple (AAPL)**.

### AAPL (Apple Inc.) Model Comparison:

The results from comparing the performance of LSTM, PyCaret, and ARIMA models for AAPL stock forecasting highlight several insights:

#### LSTM (Long Short-Term Memory):
The LSTM model showed strong performance with an **R-squared** value of **0.9956**, indicating that the model was able to explain most of the variance in the AAPL stock prices. The **MSE** value of **0.3415** and the **Adjusted R-squared** value of **0.9954** suggest that the model was a reliable predictor with minimal errors and good adjustment for overfitting.

#### PyCaret:
The PyCaret model outperformed in terms of **R-squared**, reaching an exceptional value of **0.99997**, suggesting that it fit the data almost perfectly. The **MSE** of **0.0044** and the **RMSE** of **0.0664** also highlight PyCaret's effectiveness in minimizing forecasting errors. However, it was not as interpretable as LSTM, which is a key consideration when selecting the model.

#### ARIMA (Auto-Regressive Integrated Moving Average):
The ARIMA model showed poor performance with an **R-squared** of **-0.0091**, indicating a poor fit to the data. Both **MSE** (**6.0534**) and **Adjusted R-squared** (**-0.0220**) metrics suggest that ARIMA was not a good model for AAPL stock forecasting in this instance. The model likely struggled to capture the time-series patterns in AAPL's stock price.


```python
import pandas as pd
import matplotlib.pyplot as plt

metrics_lstm = {
    'Metric': ['MSE', 'R-squared', 'Adjusted R-squared'],
    'Value': [0.341506, 0.995564, 0.995404]
}

metrics_pycaret = {
    'Metric': ['MAE', 'MSE', 'RMSE', 'R-squared'],
    'Value': [0.026100, 0.004410, 0.066409, 0.999965]
}

metrics_arima = {
    'Metric': ['MSE', 'R-squared', 'Adjusted R-squared'],
    'Value': [6.053444, -0.009076, -0.022013]
}

df_lstm = pd.DataFrame(metrics_lstm)
df_pycaret = pd.DataFrame(metrics_pycaret)
df_arima = pd.DataFrame(metrics_arima)

df_lstm.set_index('Metric', inplace=True)
df_pycaret.set_index('Metric', inplace=True)
df_arima.set_index('Metric', inplace=True)

fig, ax = plt.subplots(figsize=(10, 6))

lstm_bars = ax.bar(df_lstm.index, df_lstm['Value'], label='LSTM', width=0.2, align='center')
pycaret_bars = ax.bar(df_pycaret.index, df_pycaret['Value'], label='PyCaret', width=0.2, align='edge')
arima_bars = ax.bar(df_arima.index, df_arima['Value'], label='ARIMA', width=0.2, align='edge')

for bar in lstm_bars:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.4f}', ha='center', va='bottom')  

for bar in pycaret_bars:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.4f}', ha='center', va='bottom') 

for bar in arima_bars:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.4f}', ha='center', va='bottom') 

ax.set_title('Model Comparison: Metrics for AAPL Stock Forecasting')
ax.set_xlabel('Metric')
ax.set_ylabel('Value')
ax.legend()

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
