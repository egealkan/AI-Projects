# Web Scraping of Stock Data for MSFT and AAPL

In this notebook, we scrape historical stock price data for two major tech companies: **Microsoft (MSFT)** and **Apple (AAPL)**, using the `yfinance` library.


`yfinance` is a Python library used to download historical stock data from Yahoo Finance. It provides an easy-to-use interface to retrieve stock prices, volume, and other financial data for analysis and modeling.


```python
import yfinance as yf
```

- Importing the `yfinance` library to download stock data from Yahoo Finance: The code begins by importing the yfinance library, which is used to fetch historical stock data from Yahoo Finance.

- Downloading historical stock data for Microsoft (MSFT) from 1980-12-12 to 2024-09-01 and for Apple (AAPL) from 1980-12-12 to 2024-09-01: The code then proceeds to download the historical stock data for both Microsoft (MSFT) and Apple (AAPL). The data ranges from December 12, 1980, to September 1, 2024.

- Saving the Microsoft stock data to a CSV file named "microsoft_stock.csv" and the Apple stock data to a CSV file named "apple_stock.csv": After downloading the data, the code saves the stock data for both companies into CSV files. The Microsoft data is saved as "microsoft_stock.csv" and the Apple data as "apple_stock.csv".

- Printing a message to indicate that the data scraping process is complete: Finally, the code prints a message to the console confirming that the data scraping process has been completed successfully.



```python
dataMSFT = yf.download('MSFT', start='1980-12-12', end='2024-09-01')
dataAAPL = yf.download('AAPL', start='1980-12-12', end='2024-09-01')

dataAAPL.to_csv("apple_stock.csv")
dataMSFT.to_csv("microsoft_stock.csv")

print('Scraping data done....')
```
