## Explanation of Stock Data Columns

The dataset contains information about Apple's stock prices and trading volume. Below is an explanation of each column and the meaning of its values:

### 1. **date**
   - **Meaning:** This column represents the date on which the stock data was recorded.
   - **Values:** The values are dates, usually in `YYYY-MM-DD` format (e.g., `2024-11-27`), which represent the specific day when the stock market was open and the prices were recorded.

### 2. **adj close_aapl**
   - **Meaning:** The adjusted closing price accounts for events like stock splits, dividends, or other corporate actions that could affect the stock price. It adjusts the closing price to reflect these changes.
   - **Values:** These are the adjusted closing prices of Apple's stock for each corresponding date. For example, after a stock split, the adjusted closing price will reflect the change in the number of shares.

### 3. **close_aapl**
   - **Meaning:** The closing price is the last price at which a stock was traded during the regular trading hours (typically from 9:30 AM to 4:00 PM EST in the U.S.).
   - **Values:** These represent the closing price of stock on each specific trading day. The closing price is one of the most quoted figures in financial news reports.

### 4. **high_aapl**
   - **Meaning:** This column represents the highest price at which the stock traded during the entire trading day.
   - **Values:** These are the highest prices of the stock reached on a particular day, regardless of when during the session the peak occurred.

### 5. **low_aapl**
   - **Meaning:** This column represents the lowest price at which the stock traded during the entire trading day.
   - **Values:** These are the lowest prices that the stock reached during a specific trading session.

### 6. **open_aapl**
   - **Meaning:** The opening price is the price of the stock at the beginning of the trading day when the market opens.
   - **Values:** These represent the price of the stock when trading started on the given day (usually at 9:30 AM EST for U.S. stock markets).

### 7. **volume_aapl**
   - **Meaning:** This column represents the total number of shares of the stock that were traded during the entire trading day.
   - **Values:** These numeric values indicate the total number of shares of the stock traded on a specific day. For example, a value of `1,000,000` indicates that 1 million shares were bought and sold on that day.

---

### Example Dataset Row:
Here is an example of what one row in the dataset might look like:

| date       | adj close_aapl | close_aapl | high_aapl | low_aapl | open_aapl | volume_aapl |

---

### Summary:
- **Price columns (close, high, low, open, adj close)** represent different price points during the trading day.
- **Volume column** represents the total number of shares traded.
- **Date column** provides the timestamp for when these values were recorded.

These columns provide a detailed snapshot of the stock performance on a given day and are widely used for financial analysis, technical analysis, and stock price prediction modeling.


## Stock Data Analysis: Apple vs Microsoft

In this analysis, we are visualizing and comparing the stock price data for **Apple (AAPL)** and **Microsoft (MSFT)**. The following steps include visualizations of the closing and adjusted closing prices over time, as well as correlation heatmaps for various stock features.

### 1. **Importing Necessary Libraries**

We first import the necessary Python libraries for data manipulation and visualization.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import mplfinance as mpf


````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import mplfinance as mpf

data_aapl = 'AAPL_cleaned_data.csv'
df_aapl = pd.read_csv(data_aapl, header=[0, 1], index_col=0)

data_msft = 'MSFT_cleaned_data.csv'
df_msft = pd.read_csv(data_msft, header=[0, 1], index_col=0)

plt.figure(figsize=(12, 6))
plt.plot(df_aapl.index, df_aapl['close_aapl'], label='Closing Price', color='blue')
plt.plot(df_aapl.index, df_aapl['adj close_aapl'], label='Adjusted Closing Price', color='orange', linestyle='--')
plt.title('Apple Stock Price: Closing vs Adjusted Closing')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(df_msft.index, df_msft['close_msft'], label='Closing Price', color='green')
plt.plot(df_msft.index, df_msft['adj close_msft'], label='Adjusted Closing Price', color='red', linestyle='--')
plt.title('Microsoft Stock Price: Closing vs Adjusted Closing')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

correlation_matrix_aapl = df_aapl[['open_aapl', 'high_aapl', 'low_aapl', 'close_aapl', 'adj close_aapl', 'volume_aapl']].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix_aapl, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Apple Stock Features')
plt.tight_layout()
plt.show()

correlation_matrix_msft = df_msft[['open_msft', 'high_msft', 'low_msft', 'close_msft', 'adj close_msft', 'volume_msft']].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix_msft, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Microsoft Stock Features')
plt.tight_layout()
plt.show()
````

### Distribution of Trading Volume for Apple and Microsoft Stocks

This section of the code generates histograms to visualize the distribution of trading volumes for **Apple (AAPL)** and **Microsoft (MSFT)** stock. The goal is to observe the frequency of different trading volumes and gain insights into the activity level of both stocks over time.

#### 1. Apple Stock Trading Volume Distribution:
- **Histogram**: A histogram is created for Apple's stock trading volume (`volume_aapl`).
- **KDE Curve**: A Kernel Density Estimate (KDE) curve is added to the histogram to provide a smooth estimate of the distribution.
- **Bins**: The histogram is divided into 50 bins to represent different ranges of trading volumes.
- **Color**: The histogram bars are colored **purple** to differentiate this plot from others.


````python
plt.figure(figsize=(12, 6))
sns.histplot(df_aapl['volume_aapl'], bins=50, kde=True, color='purple')
plt.title('Distribution of Apple Stock Trading Volume')
plt.xlabel('Volume')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
sns.histplot(df_msft['volume_msft'], bins=50, kde=True, color='green')
plt.title('Distribution of Microsoft Stock Trading Volume')
plt.xlabel('Volume')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
````

### Candlestick Charts for Apple and Microsoft Stocks

This section generates **Candlestick Charts** for both **Apple (AAPL)** and **Microsoft (MSFT)** stocks to visualize price movement over time.

1. **Converting the 'date' column to datetime**:
   - The 'date' column in both dataframes is converted to a `datetime` format to properly index the stock data for visualization.


````python
df_aapl.index = pd.to_datetime(df_aapl.index)
df_msft.index = pd.to_datetime(df_msft.index)

apple_candlestick_data = df_aapl[['open_aapl', 'high_aapl', 'low_aapl', 'close_aapl']].copy()
apple_candlestick_data.columns = ['Open', 'High', 'Low', 'Close']
apple_candlestick_data = apple_candlestick_data.tail(100)

mpf.plot(apple_candlestick_data, type='candle', style='charles', title="Apple Stock Candlestick Chart", ylabel='Price ($)', figsize=(12,6))

microsoft_candlestick_data = df_msft[['open_msft', 'high_msft', 'low_msft', 'close_msft']].copy()
microsoft_candlestick_data.columns = ['Open', 'High', 'Low', 'Close']
microsoft_candlestick_data = microsoft_candlestick_data.tail(100) 

mpf.plot(microsoft_candlestick_data, type='candle', style='charles', title="Microsoft Stock Candlestick Chart", ylabel='Price ($)', figsize=(12,6))
````
