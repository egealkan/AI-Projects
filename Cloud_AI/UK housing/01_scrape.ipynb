This code is for scraping a dataset of property price records (UK housing), making a subset of the data for a specific year (2016) and saving the filtered results into a new file

```python
import pandas as pd

df = pd.read_csv('price_paid_records.csv')

# Convert 'Date of Transfer' to datetime format
df['Date of Transfer'] = pd.to_datetime(df['Date of Transfer'])

# Filter for a specific year, for example, 2017
subset_df = df[df['Date of Transfer'].dt.year == 2016]

# Save the subset to a new CSV file if needed
subset_df.to_csv('subset_2016.csv', index=False)

#Number of rows befor and after making the subset
num_rows = len(df)
print(f'Number of rows in the dataset: {num_rows}')
num_rows = len(subset_df)
print(f'Number of rows in the subset: {num_rows}')

# Display the subset dataframe
print(subset_df)
```
