### These are some visualizations before data cleaning
1. Boxplot of property prices
2. Histogram of property prices
3. Boxplot of property prices by property type

They highlight the outliers and skewed distribution of the data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv('subset_2016.csv')

# Box plot to visualize outliers
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['Price'], color='skyblue')
plt.title('Box Plot of Price (Showing Outliers)', fontsize=14)
plt.xlabel('Price', fontsize=12)
plt.show()

# Histogram for 'Price'
plt.figure(figsize=(10, 6))
sns.histplot(df['Price'], bins=30, kde=True)
plt.title('Histogram of property type')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

# Boxplot for 'Price' per 'Property Type'
plt.figure(figsize=(10, 6))
sns.boxplot(x='Property Type', y='Price', data=df)
plt.title('Price Distribution by Property Type')
plt.xlabel('Property Type')
plt.ylabel('Price')
plt.show()

```

### Visualizations after data cleaning
1. Boxplot of property prices
2. Histogram of property prices
3. Boxplot of property prices by property type

They confirm the data cleaning process and can detect remaining issues. They also support my decisions that I made while cleaning.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('cleaned_subset.csv')

# Box plot to visualize outliers
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['price'])
plt.title('Box Plot of Price (Showing Outliers)')
plt.xlabel('Price')
plt.show()

# Histogram for 'Price'
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=30, kde=True)
plt.title('Histogram of Price')
plt.xlabel('price')
plt.ylabel('Frequency')
plt.show()

property_columns = ['property_type_is__D', 'property_type_is__S', 'property_type_is__T', 'property_type_is__F']
# Add a 'property_type' column to identify the property type for each row
df['property_type'] = df[property_columns].idxmax(axis=1).str.replace('property_type_is__', '')
plt.figure(figsize=(10, 6))
sns.boxplot(x='property_type', y='price', data=df)
plt.title('Price Distribution by Property Type')
plt.xlabel('Property Type')
plt.ylabel('Price')
plt.show()
```

### Additional visualizations after cleaning
1. Distribution of property prices with mean and median line
    - This helps to identify the skewness of the data and understand how the values are spread
2. Distribution of property types
    - Helps us understand the data composition and if there are imbalances or biases

```python
hp_mean = df["price"].mean()
hp_median = df["price"].median()
sns.displot(data = df, x = 'price')
plt.axvline(x = hp_mean, color = 'blue')
plt.axvline(x = hp_median, color = "red", linestyle = '--')

property_counts = df[['property_type_is__D', 'property_type_is__S', 'property_type_is__T', 'property_type_is__F']].sum()
plt.figure(figsize=(8, 6))
property_counts.plot(kind='bar')
plt.title('Distribution of Property Types')
plt.xlabel('Property Type')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```
